**K-Nearest Neighbour (KNN) Classification**

KNN is a lazy learning algorithm.

It is the classification of data of unknown class by comparing them
with other data s in the dataframe , calculating their distance, and
assigning them to the most appropriate class according to the
calculated distance.

First, a value of k is selected.

With this k value, it is determined how many points will be evaluated
according to the nearest points around the point whose class is not
known.

Then the distance of this point with respect to the neighboring points
is calculated.

According to the calculated distances, the class of the tested data or
point is found.

Mostly,The **euclidean distance** is used when calculating the distance.

The important thing when calculating the **euclidean distance** is to do **normalization**.

Because, let’s think that , in the data, some features can be in the1000 1500 value range, while some features can be in the 0.1 value range.
If normalization is not done, data with high values dominate when calculating distance and other data seem to have no effect in the calculation.

Formula of normalization ----> **(x-min(x))/(max(x)-min(x))**

All values in column 'Unnamed: 32' appear to be missing values.
While creating the model, the 'id' and 'Unnamed: 32' columns will not be used as they will have no effect on classification

```python
data.drop(["Unnamed: 32","id"],axis=1, inplace=True)
```

```python
M = data[data.diagnosis=="M"]
B = data[data.diagnosis=="B"]
```

```python
#Classification cannot be made because the data in the diagnosis column are string values. Therefore, M=1 , B=0 values are given.
data.diagnosis = [1 if each=="M" else 0 for each in data.diagnosis] 
```

```python
y = data.diagnosis.values
```

```python
x_data= data.iloc[:,1:3].values
```

```python
#normalization
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) #n_neighbors -> k value
knn.fit(x_train,y_train)
```

```python
y_head = knn.predict(x_test)
y_head
```

```python
print("when k is {}, accuracy of knn classification {} ".format(3,knn.score(x_test,y_test)))
```

When the k value is 3, 84% of the predictions made by the k-nearest neighbor algorithm are correct.

The k value is the hyperparameter. The optimal k value is found by trying.

```python
# find k value
test_accuracy = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    test_accuracy.append(knn2.score(x_test,y_test))
    
plt.figure(figsize=(5,5))
plt.plot(range(1,15),test_accuracy)
plt.title('K Values VS Accuracy')
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.grid()
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```

```python
error = []
for i in range (1,40):
  knn= KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pred_i = knn.predict(x_test)
  error.append(np.mean(pred_i != y_test))

plt.figure(figsize=(12,6))
plt.plot(range(1,40), error, color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
plt.grid()
plt.show()
```
