**Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("flights") 
data.info
```

```python
data = sns.load_dataset("flights") 
data = data.pivot("month", "year", "passengers")

# Heatmap
sns.heatmap(data=data, annot=True, fmt="d", cmap="RdPu")

plt.show()
```

```python
data = sns.load_dataset("flights")

# Line plot
sns.lineplot(data=data, x="year", y="passengers")

plt.show()
```

```python
data = sns.load_dataset("titanic")
data.info
```

```python
data = sns.load_dataset("titanic")

# Bar plot
sns.countplot(data=data, x="embark_town")

plt.show()
```

```python
data = sns.load_dataset("titanic")

# Facet grid
g = sns.FacetGrid(data, col="sex")

g.map(sns.histplot, "age")
plt.show()
```

```python
data = sns.load_dataset("tips")
data.info
```

```python
data = sns.load_dataset("tips")
# Scatter plot
sns.scatterplot(data=data, x="total_bill", y="tip")
plt.show()
```

```python
data = sns.load_dataset("tips")

# Violin plot
sns.violinplot(x="day", y="tip", data=data)
plt.show()
```

```python
data = sns.load_dataset("iris")
data.info
```

```python
data = sns.load_dataset("iris")

# Histogram
sns.histplot(data=data, x="sepal_width")
plt.show()
```

```python
data = sns.load_dataset("iris")

# Species sütununa göre veriler gruplandırılır
g = sns.clustermap(data.drop("species", axis=1), cmap="coolwarm", standard_scale=1)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Örnek bir veri seti oluşturma
data = pd.DataFrame({
    'age_group': np.repeat(['18-25', '26-40', '41-55', '56+'], 100),
    'feedback_rating': np.random.randint(1, 6, 400)
})

# FacetGrid oluşturma
g = sns.FacetGrid(data, col="age_group", margin_titles=True)

g.map(sns.histplot, "feedback_rating")
```
