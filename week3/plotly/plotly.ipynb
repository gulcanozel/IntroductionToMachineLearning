**Plotly**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go

from google.colab import drive
drive.mount('/content/drive')
```

```python
timesData = pd.read_csv("/content/timesData.csv")
timesData
```

```python
df = timesData[timesData.year == 2016].iloc[:20,:]  #datadaki 2016 yılındaki ilk 20 üniversiteden oluşan bir filtre oluşturulup, bu filtre dataya uygulanır.
df
```

**Bar plot**

```python
trace1=go.Bar(
                    x = df.university_name,  
                    y = df.research, 
                    name = "research",  
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'), 
                    text= df.country #üstüne gelince country yazar
                 )
#trace2 
trace2=go.Bar(
                    x = df.university_name,
                    y = df.teaching,
                    name = "teaching", 
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.country
                  )
data=[trace1,trace2] 
layout = go.Layout(barmode = 'group')
fig = dict(data = data, layout = layout)
iplot(fig)
```

barplotwithplotly.png

**Scatter**

```python
trace1=go.Scatter(
                    x = df.world_rank, 
                    y = df.research, 
                    mode = "lines+markers",
                    name = "research", 
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'), 
                    text= df.university_name #(line'ın üzerine gelince üniversite ismi yazacak)
                 )
#trace2 
trace2=go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",
                    name = "teaching", 
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.university_name
                  )
data=[trace1,trace2]
layout = dict(title = 'Research and Teaching vs World Rank of Top 20 Universities in 2016',
              xaxis= dict(title= 'World Rank',ticklen= 5, zeroline= True)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

scatterlinegraph.png

```python
df1 = timesData[timesData.year == 2016].iloc[:50,:] #datadaki 2016 yılındaki ilk 50 üniversiteden oluşan bir filtre oluşturulup, bu filtre dataya uygulanır.
df2 = timesData[timesData.year == 2015].iloc[:50,:] #datadaki 2015 yılındaki ilk 50 üniversiteden oluşan bir filtre oluşturulup, bu filtre dataya uygulanır.

trace1=go.Scatter(
                    x = df2.world_rank,
                    y = df2.teaching,
                    mode = "markers",
                    name = "2015", 
                    marker={"color": "#9EA0A1"},
                    text= df2.university_name
                  )
trace2=go.Scatter(
                    x = df1.world_rank, 
                    y = df1.teaching, 
                    mode = "markers",
                    name = "2016", 
                    marker={"color": "#FFD700"}, 
                    text= df1.university_name 
                 )
#trace2 

data=[trace1,trace2]
layout = dict(title = 'Teaching vs world rank of top 50 universities with 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5, zeroline= True),
              yaxis= dict(title= 'Teaching',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

**Scatterplot Matrix**

```python
import plotly.figure_factory as ff
```

```python
#scatterplot matrix iki feature arasındaki ilişkiyi görmek için kullanılır

df3 = timesData[timesData.year == 2012] #datadaki 2012 yılından bir filtre oluşturulup, bu filtre dataya uygulanır.
data2012 = df3.loc[:,["research","citations", "teaching"]] 
data2012["index"] = np.arange(1,len(data2012)+1) 
data2012.head() #default olarak ilk beş veriyi gösterir
```

```python
fig = ff.create_scatterplotmatrix( #plotly'ın figure factory kütüphanesinden scatter plot matrix çağrılır.
                                  data2012,
                                  diag='box',  #diagonal dikdörtgenin bir köşesinden diğer köşesine çizilen çizgidir. box plot türü seçilmiştir.
                                  index='index', #index olarak oluşturulan index sütunu seçilir.
                                  colormap='Portland', #derece arttıkça maviden kırmızıya giden bir colormap
                                  colormap_type='cat',  #cat -> categorical demektir ve indeksteki her kategoriye renk haritasından bir renk atanır.
                                  height=700, width=700) 
iplot(fig)
```

**Box plot**

```python
data2013 = timesData[timesData.year == 2013] #datadaki 2013 yılından bir filtre oluşturulup, bu filtre dataya uygulanır.
```

```python
trace1 = go.Box(
    y=data2013.citations,
    name = 'citations of universities in 2013',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace2 = go.Box(
    y=data2013.research,
    name = 'research of universities in 2013',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace1, trace2]
iplot(data)
```

 **3d Scatter**

```python
data2015= timesData[timesData.year == 2015] #datadaki 2015 yılından bir filtre oluşturulup, bu filtre dataya uygulanır.

trace1 = go.Scatter3d(
    x=data2015.world_rank,  #x eksenine dünya sıralaması eklenir.
    y=data2015.teaching,    #y eksenine teaching skoru eklenir.
    z=data2015.research,   #z eksenine research skoru eklenir.
    mode='markers',          
    marker=dict(
        size=data2015.citations, #size eklenerek 4 boyutlu yapılır.
        color='rgb(321,26,0)',    
    )
)

data = [trace1]  

layout = go.Layout(
    margin=dict(  #kenarlardan bırakılan boşluktur.
        l=0,
        r=0,
        b=0,
        t=0       
    )
    
)
fig = go.Figure(data=data, layout=layout)

iplot(fig) 
```
